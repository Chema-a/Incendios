## Se detectan puntos de calor dentro del rango de 7 días y dentro de las coordenadas de la Primavera

```python
# Download a regional GeoJSON of hotspots detected by the MODIS satellite in a recent 7-day period.
def WildfireHotspots():
    data = fires.get_modis(region="central-america", time_range="7d")  
    lat = []
    lon = []
    #Detectar los incendios en el rango de 7 días dentro del rango de la primaveras
    for i in range(len(data["features"])):
        if data["features"][i]["geometry"]["coordinates"][0] >= coordinatesPrimavera[0][0] and data["features"][i]["geometry"]["coordinates"][0] <= coordinatesPrimavera[1][0] and data["features"][i]["geometry"]["coordinates"][1] <= coordinatesPrimavera[0][1] and data["features"][i]["geometry"]["coordinates"][1] >= coordinatesPrimavera[1][1]:
            lon.append(data["features"][i]["geometry"]["coordinates"][0])
            lat.append(data["features"][i]["geometry"]["coordinates"][1])
    return lat, lon


grid_points = []
grid_points.append(WildfireHotspots())
grid_points
```

```python
def distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def closest_point(p, points):
    min_distance = float('inf')
    for i, row in points.iterrows():
        dist = distance(p[1], p[0], row.geometry.centroid.x, row.geometry.centroid.y)
        if dist < min_distance:
            closest = (row.geometry.centroid.y, row.geometry.centroid.x)
            min_distance = dist
    return closest
```

```python
import geopandas as gpd
from shapely.geometry import Polygon
import numpy as np
import folium

# Carga del shapefile
gdf = gpd.read_file(r'D:\Escritorio\Modular\Primavera.shp')

# Definición de la cuadrícula
xmin, ymin, xmax, ymax = -103.6858, 20.5430, -103.4552, 20.7269
grid_size = 0.01
cols = list(np.arange(xmin, xmax, grid_size))
rows = list(np.arange(ymin, ymax, grid_size))
rows.reverse()

polygons = []
for x in cols:
    for y in rows:
        polygons.append(Polygon([(x,y), (x+grid_size,y), (x+grid_size,y-grid_size), (x,y-grid_size)]))

grid = gpd.GeoDataFrame({'geometry':polygons})
intersected = gpd.overlay(gdf, grid, how='intersection')

# Calcular los puntos más cercanos a los polígonos de la cuadrícula
closest_points = []
for p in grid_points:
    closest_points.append(closest_point(p, intersected))


# Creación de un mapa de folium con el shapefile y la cuadrícula
m = folium.Map(location=[20.6, -103.57], zoom_start=12)

# Agregar shapefile al mapa
folium.GeoJson(gdf).add_to(m)

for index, row in intersected.iterrows():
    tooltip = f"{index}"
    y_centrum, x_centrum = row.geometry.centroid.y, row.geometry.centroid.x
    if (y_centrum, x_centrum) in closest_points:
        marker =  folium.Marker([y_centrum, x_centrum], 
                  icon=folium.Icon(color='red', icon='pushpin'))
    else:
        marker = folium.Marker([y_centrum, x_centrum], tooltip=tooltip)
    marker.add_to(m)
    

# Mostrar el mapa
m
    
```

## Recolección de datos atmosféricos de cada cuadrante

```python
# Definir la URL de la API de OpenWeatherMap y los parámetros de consulta
url = "https://api.openweathermap.org/data/2.5/weather"

#Definimos las listas de las variables
wind_speed = []
temp = []
wind_direction = []
slope = []
fuel_moisture = []


for i, row in intersected.iterrows():
    xmin, ymin, xmax, ymax = row.geometry.bounds
    
    x_centrum = (xmin + xmax)/2 
    y_centrum = (ymin + ymax)/2 

    print(f"Coordenadas del cuadrante {i}: {x_centrum},{y_centrum}")
    
    params = {
        "lat": y_centrum, # Latitud
        "lon": x_centrum, # Longitud
        "appid": "88ea1e1a48e480178b9c4ab177c2475f", # Tu ID de API de OpenWeatherMap
        "units": "metric" # Unidades métricas (para obtener la velocidad del viento en m/s)
    }
    # Hacer una solicitud HTTP a la API y obtener la respuesta JSON
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
    
        # Extraer la velocidad del viento
        wind_speed.append(data["wind"]["speed"])
        temp.append(data["main"]["temp"])
        wind_direction.append(data["wind"]["deg"])

        # Extraer la altitud y la presión atmosférica para calculo de pendiente de la respuesta JSON
        altitude = None
        atmosferic_pressure = data["main"]["pressure"] # En hPa
        try:
            altitude = data["main"]["sea_level"] # En metros
            slope.append(0.102 * altitude / atmosferic_pressure) # En grados
        except KeyError:
            slope.append(0)
        # Calcular la pendiente


        # Extraer los valores de temperatura, humedad relativa, velocidad del viento y precipitación de la respuesta JSON
        temp_calculate = data["main"]["temp"] - 273.15 # Convertir de Kelvin a Celsius
        relative_moisture = data["main"]["humidity"]
        precipitation = data["rain"]["1h"] if "rain" in data else 0 # Obtener la precipitación de los últimos 60 minutos (si está disponible)
        wind_speed_calculate = data["wind"]["speed"]

        # Calcular la humedad de combustible utilizando la fórmula de Deeming
        fuel_moisture.append(20 + 280 / (temp_calculate + 273) - relative_moisture + 0.1 * wind_speed_calculate + 0.2 * precipitation)
    else:
        print('La solicitud falló con el código de estado:', response.status_code)
    
# Imprimir los resultados
print(f"Velocidad del viento: {wind_speed} m/s")
print(f"Pendiente: {slope} grados")
print(f"Humedad del combustible: {fuel_moisture}%")
print("Temperatura: ", temp,"°C")
print("Direccion del viento:",wind_direction,"°")
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
# Constantes para el modelo de Rothermel
c1 = 0.0579
c2 = 0.0651
c3 = 0.0231
c4 = 0.0115
p1 = 0.2427
p2 = 0.1528
p3 = 0.0182
S = 0.0555

# Función para calcular el índice de propagación del fuego (Rothermel)
def rothermel_model(slope, wind_speed, fuel_moisture):
    # Convertir de km/h a m/s
    wind_speed = wind_speed / 3.6
    
    # Calcular las constantes auxiliares
    sigma = 1 / (1 + 4.88 / fuel_moisture)
    gamma = 0.0015 * fuel_moisture
    
    # Calcular los coeficientes de propagación
    c = c1 + c2 * sigma + c3 * slope + c4 * gamma
    e = (0.936 * sigma + 0.256) / (1 + 1.9 * sigma)
    f = 1.5 + 0.25 * sigma
    n = (p1 * wind_speed + p2) * (1 - math.exp(-p3 * fuel_moisture)) * f
    phi = math.atan(slope / 100)
    if np.isnan(fuel_moisture) or np.isnan(wind_speed):
        beta = 0  # Asignar un valor predeterminado o tomar alguna otra acción adecuada
    else:
        beta = math.atan((S + 0.05 * wind_speed) / (1.5 * abs(fuel_moisture) ** 0.5 * math.cos(phi)))
    delta = 1 + wind_speed / 7
    
    # Calcular el índice de propagación del fuego (Rothermel)
    I = c * e * n * delta * math.cos(beta)
    
    return I

# Crear un DataFrame para almacenar los resultados por intervalo de tiempo
results_df = pd.DataFrame(columns=["Cuadrante", "Intervalo", "Indice_Propagacion"])

# Calcular el índice de propagación del fuego para cada cuadrante y intervalo de tiempo
for interval in [1, 3, 9]:  # Intervalos de tiempo en horas
    I = []
    for i, row in intersected.iterrows():
        slope_val = slope[i] # Pendiente en grados
        wind_speed_val = wind_speed[i] # Velocidad del viento en m/s
        fuel_moisture_val = fuel_moisture[i] # Humedad de combustible
        I_val = rothermel_model(slope_val, wind_speed_val, fuel_moisture_val) # Índice de propagación del fuego
        I.append(I_val)
        
    # Agregar resultados al DataFrame
    results_df = results_df.append({"Cuadrante": i, "Intervalo": interval, "Indice_Propagacion": I}, ignore_index=True)
    print(results_df)

# Merge del shapefile con los resultados
intersected = gdf.merge(results_df, left_index=True, right_on="Cuadrante")

```
